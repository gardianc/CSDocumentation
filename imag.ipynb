```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, triang
```

### triangular

```python
# --- 1. Configuración de parámetros ---
# Definimos valores de ejemplo para a y b para hacer el gráfico
b,d = 0.1,1.1
x = np.linspace(b, d, 1000)
c = 0.5
dist = triang(c, loc=b, scale=1)
y = dist.pdf(x)
# Calculamos mu y sigma basados en a y b
mu_val = dist.median()
# La desviación estándar de una uniforme es (b-a) / sqrt(12)
sigma_val = dist.std()

# Definimos los puntos clave para el eje x
x_puntos = {
    r'$b$':b,
    r'$\mu-\sigma$': mu_val - sigma_val,
    r'$\mu$': mu_val,
    r'$\mu+\sigma$': mu_val + sigma_val,
    r'$d$': d
}

print(dist.pdf(mu_val),1/(mu_val-b),2/(d-b))

plt.rcdefaults()
# --- 2. Crear la figura y los ejes ---
fig, ax = plt.subplots(figsize=(8, 4.5))

# --- 3. Dibujar la distribución (el rectángulo) ---
# Dibujamos las líneas sólidas exteriores
ax.plot(x, y, color='black', linewidth=1.5)  # Curva de la distribución
# --- 4. Dibujar líneas verticales interiores ---
# Línea sólida en mu
ax.vlines(x=mu_val, ymin=0, ymax=dist.pdf(mu_val), color='black', linestyle='solid', linewidth=1)
# Líneas punteadas en mu-sigma y mu+sigma
ax.vlines(x=x_puntos['$\mu-\sigma$'], ymin=0, ymax=dist.pdf(mu_val-sigma_val), color='black', linestyle='dashed', linewidth=1)
ax.vlines(x=x_puntos['$\mu+\sigma$'], ymin=0, ymax=dist.pdf(mu_val+sigma_val), color='black', linestyle='dashed', linewidth=1)

ax.hlines(y = 1/c, xmin = 0, xmax = mu_val, color='black', linestyle='dashed', linewidth=1)

# --- 6. Configuración de Ejes y Etiquetas ---
# Usar flechas para los ejes
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')

# --- 7. Ticks y etiquetas de texto ---
ax.set_xlim(0)
# Eje X
# Configuramos la posición de los ticks y sus etiquetas de texto
ax.set_xticks(list(x_puntos.values()))
ax.set_xticklabels(list(x_puntos.keys()), fontsize=12)

ax.set_yticks([1/c])
# Usamos LaTeX para la fracción del tick del eje y
ax.set_yticklabels([r'$\frac{2}{d-b}$'], fontsize=16)

# Mostrar el gráfico
# plt.tight_layout()
plt.show()
```

### rectangular

```python
# --- 1. Configuración de parámetros ---
# Definimos valores de ejemplo para a y b para hacer el gráfico
a_val = 1.5
b_val = 8.0

# Calculamos mu y sigma basados en a y b
mu_val = (a_val + b_val) / 2
# La desviación estándar de una uniforme es (b-a) / sqrt(12)
sigma_val = (b_val - a_val) / np.sqrt(12)

# Definimos los puntos clave para el eje x
x_puntos = {
    '0': 0,
    'a': a_val,
    r'$\mu-\sigma$': mu_val - sigma_val,
    r'$\mu$': mu_val,
    r'$\mu+\sigma$': mu_val + sigma_val,
    'b': b_val
}

plt.rcdefaults()
# --- 2. Crear la figura y los ejes ---
fig, ax = plt.subplots(figsize=(8, 4.5))

# --- 3. Dibujar la distribución (el rectángulo) ---
# Dibujamos las líneas sólidas exteriores
ax.plot([a_val, b_val], [1, 1], color='black', linewidth=1.5)  # Techo
ax.plot([a_val, a_val], [0, 1], color='black', linewidth=1.5)  # Pared izquierda
ax.plot([b_val, b_val], [0, 1], color='black', linewidth=1.5)  # Pared derecha

# --- 4. Dibujar líneas verticales interiores ---
# Línea sólida en mu
ax.vlines(x=mu_val, ymin=0, ymax=1, color='black', linestyle='solid', linewidth=1)
# Líneas punteadas en mu-sigma y mu+sigma
ax.vlines(x=x_puntos['$\mu-\sigma$'], ymin=0, ymax=1, color='black', linestyle='dashed', linewidth=1)
ax.vlines(x=x_puntos['$\mu+\sigma$'], ymin=0, ymax=1, color='black', linestyle='dashed', linewidth=1)

# --- 5. Dibujar línea de proyección al eje Y ---
# Proyección punteada desde el punto 'a' al eje Y
ax.plot([0, a_val], [1, 1], color='black', linestyle='dashed', linewidth=1)

# --- 6. Configuración de Ejes y Etiquetas ---
# Usar flechas para los ejes
ax.spines['left'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')

# --- 7. Ticks y etiquetas de texto ---

# Eje X
# Configuramos la posición de los ticks y sus etiquetas de texto
ax.set_xticks(list(x_puntos.values()))
ax.set_xticklabels(list(x_puntos.keys()), fontsize=12)

# Eje Y
# Solo queremos un tick para la altura del rectángulo
ax.set_yticks([1])
# Usamos LaTeX para la fracción del tick del eje y
ax.set_yticklabels([r'$\frac{1}{b-a}$'], fontsize=14)
# Mover la etiqueta un poco a la izquierda para que se vea bien
# ax.tick_params(axis='y', pad=10)

# --- 8. Límites del gráfico ---
# Ajustar los límites para que se vea el 0 y haya espacio al final de la flecha
ax.set_xlim(0, b_val + 1.5)
ax.set_ylim(0, 1.3)

# --- 9. Limpieza visual ---
# Quitar los valores numéricos predeterminados de los ejes y
ax.set_ylabel('')
ax.set_xlabel('')

# Mostrar el gráfico
# plt.tight_layout()
plt.show()
```
